Step 1: Loading all the libraries and datasets. 

```python
#Importing the libraries. 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn  
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, mean_absolute_error 
from sklearn.metrics import r2_score 
```

```python
#Removing the skewness - making the data normal/ 
#0 -> Log transfomration
from scipy.stats import boxcox 
btdf['CRIM'] = boxcox(btdf['CRIM'], 0)
```

```python
sns.distplot(btdf['CRIM'])
```

```python
#Removing outliers from the dataset. 

import scipy.stats 
from scipy.stats import zscore 
z = np.abs(zscore(btdf))
```

```python
threshold = 3 
print(np.where(z>3))
```

```python
btdf_new = btdf[(z<3).all(axis = 1)]
```

```python
# The data with normal data values. 
btdf = btdf_new
```

```python
btdf
```

```python
sns.pairplot(btdf)
```

Step 3: Splitting the data into training and test phase. Since the data btdf is cleaned now. 

```python
x = btdf.iloc[:,0:13]
y = btdf.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = 45)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Step 4: Model selection and validation

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
btdf.columns
```

```python
lm.score(x_train, y_train)
```

```python
#Predicting the results. 
pred = lm.predict(x_test)
print("Predicted result price: ", pred)
print("Actual result price: ", y_test)
```

```python
print("ERROR!")
print('Mean absolute error: ', mean_absolute_error(y_test, pred))
print('Mean squared error: ', mean_squared_error(y_test, pred))
```

```python
#Coefficnet of determination 

print(r2_score(y_test,pred))
```
